# Data Exploration

```python
import pandas as pd
import numpy as np
```

```python
# Read the dataset into a pandas DataFrame
data = pd.read_csv('../data/cleaned/cleaned_data.csv')
display(data)
```

## Research Question 1: How has the frequency of terrorist attacks changed over time?

```python
attacks_by_year = data.groupby('year').size()
attacks_by_year_df = pd.DataFrame(attacks_by_year, columns=['Number of Attacks'])
display(attacks_by_year_df)
```

## Research Question 2: Which countries have experienced the highest number of terrorist incidents?

```python
incidents_by_country = data['country'].value_counts()
print(incidents_by_country.head())
```

## Research Question 3: What were the most common types of terrorist attacks?

```python
common_attack_types = data['attack_type'].value_counts()
print(common_attack_types)
```

## Research Question 4: How have terrorist attack varied by weapon type across regions?

```python
attacks_by_region_weapon = data.groupby(['region', 'weapon_type']).size()
display(attacks_by_region_weapon)
```

## Research Question 5: What Are the Most Common Target Types in Terrorist Attacks?

```python
common_target_types = data['target_type'].value_counts()
print(common_target_types)
```
